## Load csv and create two seperate dataframe:
We are going to use heart dataframe to differentiate healty people from the people with diaseaes. Data's of healty people will be stored in the healty dataset while diseased people will be in diseased dataset.

```python
heart = pd.read_excel("/Users/ecem/Desktop/seismocardiogram/data/diseased-dataset/Summary_Pub_Deidentified.xlsx")
df = heart.copy().fillna(0)
```

I have dropped some columns because they didnt have much information. ????????

```python
df.drop(["Mitral valve Peak velocity (m/sec)",
         "Mitral valve Mean gradient(mmHg)", 
         "Echo available",
        "Date of collection",
        "Date of echo"], 
        inplace = True, axis = 1)
df.drop(df.tail(1).index,inplace=True)
```

## only without TR

```python
idx = np.where(df['moderate or greater TR'] != 1.0)
```

```python
idx
```

```python
idx_  = np.where((df['Moderate or greater MS'] == 0.0 ) &
              (df['Moderate or greater MR'] == 0.0 )&
              (df['Moderate or greater AR'] == 0.0 )&
              (df['Moderate or greater AS'] == 0.0))
idx_[0]
```

```python
final_idx = list(set(idx[0]) - set(idx_[0]))
```

```python
final_idx
```

```python
len(final_idx)
```

```python
diseased = df.loc[final_idx]
```

```python
diseased
```

```python
diseased_x, diseased_y, diseased_z = [],[],[]
```

```python
for i in range(len(diseased['Patient ID'].index)):
    x_ = x[diseased['Patient ID'].index[i]]
    y_ = y[diseased['Patient ID'].index[i]]
    z_ = z[diseased['Patient ID'].index[i]]
    
    diseased_x.append(x_)
    diseased_y.append(y_) 
    diseased_z.append(z_) 
del x_, y_, z_
```

```python
stacked_diseased = np.stack((diseased_x, diseased_y, diseased_z), axis = -1)
```

```python
print(stacked_diseased.shape)
```

```python
np.save(path + "/diseased_gyro", stacked_diseased, allow_pickle= True)
```
