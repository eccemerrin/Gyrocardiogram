```python
ms_ = np.load(dir_path + "ms-10.npy", allow_pickle= True)
mr_ = np.load(dir_path + "mr-10.npy", allow_pickle= True)
as_ = np.load(dir_path + "as-10.npy", allow_pickle= True)
ar_ = np.load(dir_path + "ar-10.npy", allow_pickle= True)
```

```python
def plotter(array, disease = "mr", num_of_it = 10):
    idx = np.random.randint(0, array.shape[1], size = num_of_it)
    for i in idx:
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (18,4))
        ax1.plot(array[0][i], color = 'pink')
        ax1.title.set_text('x comp. of {}th {} gyro '.format(i, disease))
        
        ax2.plot(array[1][i], color = 'purple')
        ax2.title.set_text('y comp. of {}th {} gyro '.format(i, disease))
        
        ax3.plot(array[2][i], color = 'orange')
        ax3.title.set_text('z comp. of {}th {} gyro '.format(i, disease))
```

```python
plotter(mr_, disease = "mr")
```

```python
plotter(ms_, disease = "ms")
```

```python
plotter(ar_, disease = "ar")
```

```python
plotter(as_, disease = "as")
```
