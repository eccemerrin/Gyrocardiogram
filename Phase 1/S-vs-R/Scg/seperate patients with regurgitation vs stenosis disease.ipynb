## Load csv and create two seperate dataframe:
We are going to use heart dataframe to differentiate healty people from the people with diaseaes. Data's of healty people will be stored in the healty dataset while diseased people will be in diseased dataset.

```python
heart = pd.read_excel("/Users/ecem/Desktop/seismocardiogram/data/diseased-dataset/Summary_Pub_Deidentified.xlsx")
df = heart.copy().fillna(0)
```

I have dropped some columns because they didnt have much information. ????????

```python
df.drop(["Mitral valve Peak velocity (m/sec)",
         "Mitral valve Mean gradient(mmHg)", 
         "Echo available",
        "Date of collection",
        "Date of echo"], 
        inplace = True, axis = 1)
df.drop(df.tail(1).index,inplace=True)
```

## only without TR

```python
idx = np.where(df['moderate or greater TR'] != 1.0)
```

```python
idx
```

## only with s

```python
idx_s  = np.where(((df['Moderate or greater MS'] == 1.0 ) | (df['Moderate or greater AS'] == 1.0))
         & (df['Moderate or greater MR'] == 0.0 )
         & (df['Moderate or greater AR'] == 0.0 ))
```

```python
idx_s
```

```python
idx_s[0].shape
```

## only with regurgitation

```python
idx_r  = np.where((df['Moderate or greater MS'] == 0.0 ) 
            &((df['Moderate or greater MR'] == 1.0 )| (df['Moderate or greater AR'] == 1.0 )) 
            &(df['Moderate or greater AS'] == 0.0))
idx_r[0]
```

```python
idx_r[0].shape
```

## split the dataframe

```python
diseased_s = df.loc[idx_s]
```

```python
diseased_s
```

```python
diseased_r = df.loc[idx_r]
diseased_r
```

```python
diseased_x_s, diseased_y_s, diseased_z_s = [],[],[]
diseased_x_r, diseased_y_r, diseased_z_r = [],[],[]
```

```python
for i in range(len(diseased_s['Patient ID'].index)):
    x_ = x[diseased_s['Patient ID'].index[i]]
    y_ = y[diseased_s['Patient ID'].index[i]]
    z_ = z[diseased_s['Patient ID'].index[i]]
    
    diseased_x_s.append(x_)
    diseased_y_s.append(y_) 
    diseased_z_s.append(z_) 
del x_, y_, z_
```

```python
stacked_diseased_s = np.stack((diseased_x_s, diseased_y_s, diseased_z_s), axis = -1)
```

```python
print(stacked_diseased_s.shape)
```

```python
for i in range(len(diseased_r['Patient ID'].index)):
    x_ = x[diseased_r['Patient ID'].index[i]]
    y_ = y[diseased_r['Patient ID'].index[i]]
    z_ = z[diseased_r['Patient ID'].index[i]]
    
    diseased_x_r.append(x_)
    diseased_y_r.append(y_) 
    diseased_z_r.append(z_) 
del x_, y_, z_
```

```python
stacked_diseased_r = np.stack((diseased_x_r, diseased_y_r, diseased_z_r), axis = -1)
```

```python
print(stacked_diseased_r.shape)
```
