```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_path = "/Users/ecem/Desktop/gyrocardiogram/data/"
```

```python
ms_ = np.load(dir_path + "ms.npy", allow_pickle= True)
mr_ = np.load(dir_path + "mr.npy", allow_pickle= True)
as_ = np.load(dir_path + "as.npy", allow_pickle= True)
ar_ = np.load(dir_path + "ar.npy", allow_pickle= True)
```

```python
print(ms_.shape)
print(mr_.shape)
print(as_.shape)
print(ar_.shape)
```

```python
def plotter(array, disease = "mr"):
    for i in range(array.shape[1]):
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (18,4))
        ax1.plot(array[i][0], color = 'pink')
        ax1.title.set_text('x comp. of {}th {} gyro '.format(i, disease))
        
        ax2.plot(array[i][1], color = 'purple')
        ax2.title.set_text('y comp. of {}th {} gyro '.format(i, disease))
        
        ax3.plot(array[i][2], color = 'orange')
        ax3.title.set_text('z comp. of {}th {} gyro '.format(i, disease))
```

```python
plotter(mr_, disease = "mr")
```

```python
plotter(ms_, disease = "ms")
```

```python
plotter(ar_, disease = "ar")
```

```python
plotter(as_, disease = "as")
```
